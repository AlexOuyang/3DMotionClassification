```python
def read_in_data():
    data_path = "../Data/Gestures/a.txt"
    
    with open("../Data/Gestures/a.txt") as dataFile:
        data = dataFile.readlines()
        json_array = []
        
        for line in data:
            json_group = json.loads(line)
            for curr_json in json_group['data']:
                json_array.append(curr_json)
        result = json_normalize(json_array)
        return result
```
